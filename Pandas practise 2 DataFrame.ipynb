DataFrame保存的数据为二维数组形式，其可以认为每一行都是一个样本，每一列都是样本的一种属性

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from pandas import DataFrame
```

```python
df = DataFrame(data = {'Python':np.random.normal(loc = 120, scale = 5, size = 18).round(1),
                       'Math':np.random.normal(loc = 110, scale = 15, size = 18).round(1),
                      'English':np.random.normal(loc = 110, scale = 30, size = 18).round(1)},
              index = list('ABCDEFGHIJKLMNOPQR'))
df
```

```python
df2 = DataFrame(data = np.random.randint(0,150,size = (10,4)),
                index = list('ABCDEFGHIJ'),
                columns = ['Python','Math','English','Physic'])

df2
```

```python
df2.to_csv('./data.txt')
```

```python
df2.to_excel('./data.xlsx')
```

```python
df3 = pd.read_csv('./data.txt')
df4 = df3.rename(mapper = {'Unnamed: 0':'index'},axis = 1)
```
