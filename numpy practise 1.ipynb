级联
np.concatenate()级联需要注意：
1、级联的参数是列表，一定要加中括号或者小括号
2、维度必须相同
3、形状相符
*****级联的方向默认是shape这个tuple的第一个值所代表的维度方向
可以通过axis参数改变级联的方向

```python
xz_data_4 = xz_data[:,:670]
plt.imshow(xz_data_4)
```

```python
xz_data_5 = xz_data[:,671:,::-1]
plt.imshow(xz_data_5)
```

```python
np.concatenate([xz_data_4,xz_data_5],axis = 1)
```

```python
plt.imshow(np.concatenate([xz_data_4,xz_data_5],axis = 1))
```
