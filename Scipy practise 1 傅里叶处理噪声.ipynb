```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from scipy.io import wavfile
from scipy.fftpack import fft2,ifft2

%matplotlib inline

sp.__version__
```

```python
xz = plt.imread('./timg.jpg')
plt.figure(figsize = (16,12))
plt.imshow(xz)
```

```python
xz_fft = fft2(xz)
conditions = 12* (np.abs(xz_fft).mean())
cond = np.abs(xz_fft) > conditions

cond
```

```python
xz_fft[cond] = 0
xz_fft
```

```python
xz_result = ifft2(xz_fft)
xz2 = np.real(xz_result)
xz2.shape
```

```python
plt.figure(figsize = (12,8))
plt.subplot(2,2,1)
plt.imshow(xz2[:,:,0], cmap = plt.cm.gray)

plt.subplot(2,2,2)
plt.imshow(xz2)

plt.subplot(2,2,3)
plt.imshow(xz)

plt.subplot(2,2,4)
plt.imshow(xz[:,:,0], cmap = 'gray')
```

```python
xz3 = np.abs(xz2[:,:,0]).astype(np.uint8)

xz3_w = xz3 + 20
xz3_b = xz3 - 100
```

```python
plt.figure(figsize = (12,8))
plt.subplot(2,2,1)
plt.imshow(xz3)

plt.subplot(2,2,2)
plt.imshow(xz3,cmap = plt.cm.gray)

plt.subplot(2,2,3)
plt.imshow(xz3_w,cmap = plt.cm.gray)

plt.subplot(2,2,4)
plt.imshow(xz3_b,cmap = plt.cm.gray)
```
