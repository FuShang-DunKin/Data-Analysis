```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pandas import Series, DataFrame
```

pandas 提供了两种拼接方式

级联，pandas.concat, pandas.append

合并，pandas.merge, pandas.join

```python
# numpy的级联
nd1 = np.random.randint(0, 150, size = (4, 5))

nd2 = np.random.randint(0, 150, size = (7, 5))

nd3 = np.random.randint(0, 150, size = (4, 8))
```

```python
np.concatenate([nd1, nd2])           #默认axis = 0
```

```python
# pandas 的级联和numpy是类似的
df1 = DataFrame(np.random.randint(0, 150, size = (5, 3)),
                columns = ['Python', ' English', 'Math'],
                index = list('ABCDE'))

df2 = DataFrame(np.random.randint(0, 150, size = (5,3)),
               columns = ['Python', ' English', 'Math'],
               index = list('HIJKL'))
display(df1, df2)
```

```python
#一般级联，通常是在行上的级联，因为一般不同的表进行级联的前提是具有相同的属性
pd.concat([df1, df2])
```

```python
df3 = DataFrame(np.random.randint(0, 150, size = (5, 3)),
               columns = ['Chiese', 'Physic', 'Chem'],
               index = list('OPQRS'))
df3
```

```python
df4 = pd.concat([df1, df3], axis = 1)
df4.fillna(0)
```

concat进行级联的时候，参数join默认是outer即合并所有，改为inner只对共同元素进行合并

merge 与 concat的区别在于，merge需要依据某一共同的行或列来进行合并

使用pandas.merge()合并时，会自动根据两者相同的columns名称的那一列，作为key来进行合并

注：每一列元素的顺序不要求一致

```python
# 一对一的合并
df1 = DataFrame({'id':[1, 10, 1024], 'name': ['po', 'alisa', 'Michael'], 'sex': ['male', 'female', 'male']})

df2 = DataFrame({'id': [1, 11, 1025], 'salary': [10000,20000,30000], 'age': [20,30,40]})

display(pd.concat([df1, df2]), df1.merge(df2))
```

```python
# 多对一的合并
df1 = DataFrame({'id':[1, 10, 1024], 'name': ['po', 'alisa', 'Michael'], 'sex': ['male', 'female', 'male']})

df2 = DataFrame({'id': [1, 1, 1], 'salary': [10000,20000,30000], 'age': [20,30,40]})

df1.merge(df2)
```

```python
# 多对多合并
df3 = DataFrame({'id':[1, 1, 1], 'name': ['po', 'alisa', 'Michael'], 'sex': ['male', 'female', 'male']})

pd.merge(df2,df3)
```

key的规范化使用

使用on= 显示指定的那一列作为key，当有多个key相同的时候使用

```python
df1 = DataFrame({'id':[1, 10, 1024], 'name': ['po', 'alisa', 'Michael'], 'sex': ['male', 'female', 'male']})

df2 = DataFrame({'id': [1, 11, 1025], 'name': ['softpo', 'alisa', 'Michael'], 'age': [20,30,40]})

display(df1, df2)

pd.merge(df1, df2, on = 'name')
```
