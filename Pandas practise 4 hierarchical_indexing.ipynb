数据分组处理








核心是gruopby()函数

```python
df = DataFrame({
    'item': np.random.randint(0, 10, size = 100),
    'sailer': np.random.randint(0, 10, size = 100),
    'weight': np.random.randint(30, 300, size = 100),
    'price': np.random.randint(1, 20, size = 100)
})
df
```

```python
def convert_item(x):
    if x < 2 :
        return '萝卜'
    elif x < 7 :
        return '白菜'
    else :
        return '青椒'

df['item'] = df['item'].map(convert_item)
```

```python
df
```

```python
def convert_sailer(x):
    if x < 5:
        return '张大妈'
    elif x < 7:
        return '王大妈'
    else :
        return '于大爷'

df['sailer'] = df['sailer'].map(convert_sailer)

df
```

```python
def convert_weight(x):
    if x < 51 :
        return 50
    elif x < 150 :
        return 100
    else :
        return 200
    
df['weight'] = df['weight'].map(convert_weight)

df
```

```python
# 统计这一百天三位买菜员的销售总额
ret = df.groupby(['sailer'])['weight']
ret
```

```python
# 分组后进行计算
ret.sum()
```

```python
ret.min()
```

```python
ret.apply(np.sum)
```

```python
ret.count()
```

```python
ret.apply(np.size)
```
