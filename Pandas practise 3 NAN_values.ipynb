```python
# 空数据处理，属于数据清洗的一部分,即数据预处理的一个方面
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from pandas import Series,DataFrame
```

```python
df = DataFrame(np.random.randint(0,150,size = (150,3)), columns = ['Python','Math','English'])
df
```

```python
for i in range(15):
    index = np.random.randint(0,150,size = 1)[0]
    column = np.random.randint(0,3,size = 1)[0]
    df.iloc[index,column] = None
```

```python
df
```

```python
df.info()
```

处理空数据第一步，定义空数据

```python
cond = df.isnull().any(axis = 1)
df[cond]
```

处理空数据，当空数据不多时，直接删除

```python
# 非空数据
cond = df.notnull().all(axis = 1)

df[cond]
```

```python
# 删除空数据
df.dropna()
```

```python
# 按指定索引删除空数据
cond = df.isnull().any(axis = 1)
index = df[cond].index

df1 = df.drop(labels = index)
```

同理可以进行数值过滤，清洗掉较小、较大、或均值不满足条件的数据 

```python
cond = (df1 < 60).any(axis = 1)

index = df1[cond].index

df1.drop(labels = index)
```

```python
cond = (df1.mean(axis = 1) < 90)

index = df1[cond].index

df1.drop(labels = index)
```

```python
cond1 = df1.mean(axis = 1) < 60
cond2 = df1.mean(axis = 1) > 120

cond = cond1 | cond2

df1[cond]
```

#np.nan
#np.nan是浮点类型的数，能够参与到计算中，但是计算结果总是NaN

可以使用np.nan*()函数计算NaN，此时NaN取值为0参与计算

```python
df = DataFrame(np.random.randint(0,150,size = (6,3)), columns = ['Python','English', 'Math'])
```

```python
df['Python'][0] = None

df['English'][2] = np.NaN

df
```

使用“假数据”对空数据进行填充时，应尽量使用更为合理的数值代替空值

```python
df.fillna(value = df.mean())
```
