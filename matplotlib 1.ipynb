```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,2*np.pi)
y = np.sin(x)

plt.plot(x,y)
plt.grid()
```

```python
plt.figure(figsize = (12,3))
ax = plt.subplot(1,3,1)
ax.plot(x,y)
ax.grid()

ax = plt.subplot(1,3,2)
ax.plot(x,y)
ax.grid(color = 'red',alpha = 0.8)

ax = plt.subplot(1,3,3)
ax.plot(x,y)
ax.grid(color = 'green', linestyle = '-.')
```

```python
plt.plot(x,y)
plt.axis([-2,8,-1,1])
```

```python
plt.subplot(1,2,1)
line = plt.plot(x,y)
line[0].set_color('red')

plt.subplot(1,2,2)
line, = plt.plot(x,y)
line.set_color('green')
```

```python
angle = np.linspace(0,2*np.pi,200)

x = np.sin(angle)
y = np.cos(angle)

# plt.figure(figsize = (5,5))
# plt.axis('equal')
plt.plot(x,y)
plt.xlim([-2,2])
plt.ylim([-2,2])
```
