```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt

from scipy import integrate
from scipy import io
from scipy.fftpack import fft2,ifft2
from scipy import misc
from scipy import ndimage

%matplotlib inline
```

```python
circle = lambda x:(1-x**2)**0.5

x = np.linspace(-1,1,200)


plt.figure(figsize = (9, 9))
plt.plot(x, circle(x))
plt.plot(x, -circle(x))
```

```python
integrate.quad(circle,-1,1)[0] * 2
```

```python
xz = plt.imread('./timg.jpg')
xz_fft = fft2(xz)
conditions = 18* (np.abs(xz_fft).mean())
cond = np.abs(xz_fft) > conditions
xz_fft[cond] = 0
xz_result = ifft2(xz_fft)
xz2 = np.real(xz_result)
xz2 = np.real(xz2)

io.savemat('xz.mat', mdict = {'xz':xz2[:,:,2]})
```
